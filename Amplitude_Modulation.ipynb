```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

Ac = float(input("Enter carrier amplitude = "))  # carrier amplitude
fc = float(input("Enter carrier frequency = "))  # carrier frequency
Am = float(input("Enter message signal amplitude = "))  # message signal amplitude
fm = float(input("Enter message signal frequency = "))  # message signal frequency
Fs = 100  # sampling rate/frequency

ka = 1  # amplitude sensitivity
t = np.arange(0, 50, 0.1)  # defining the time range and discretizing it into samples
ct = Ac * np.sin(2 * np.pi * fc * t)  # defining the carrier signal wave
mt = Am * np.cos(2 * np.pi * fm * t)  # defining the message signal wave
AM = ct * (1 + ka * mt)  # amplitude modulated wave, according to the standard definition

plt.figure(figsize=(10, 10))

plt.subplot(3, 1, 1)  # plotting the message signal wave
plt.plot(t, mt)
plt.title('Message signal')
plt.xlabel('Time')
plt.ylabel('Messaging signal')

plt.subplot(3, 1, 2)  # plotting the carrier signal wave
plt.plot(t, ct)
plt.title('Carrier Signal')
plt.xlabel('Time')
plt.ylabel('Carrier')

plt.subplot(3, 1, 3)  # plotting the amplitude modulated wave
plt.plot(t, AM)
plt.title('Amplitude Modulated Signal')
plt.xlabel('Time')
plt.ylabel('AM signal')

plt.tight_layout()
plt.show()
```
