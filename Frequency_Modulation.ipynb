```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 1, 0.001)  # up to 1000 samples
vm = float(input('Enter Amplitude (Message) = '))
vc = float(input('Enter Amplitude (Carrier) = '))
fM = float(input('Enter Message frequency = '))
fc = float(input('Enter Carrier frequency = '))
m = float(input('Enter Modulation Index = '))
msg = vm * np.sin(2 * np.pi * fM * t)
plt.subplot(3, 1, 1)  # plotting message signal
plt.plot(t, msg)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Message ')

carrier = vc * np.sin(2 * np.pi * fc * t)
plt.subplot(3, 1, 2)  # plotting carrier signal
plt.plot(t, carrier)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Carrier Signal')

y = vc * np.sin(2 * np.pi * fc * t + m * np.cos(2 * np.pi * fM * t))
plt.subplot(3, 1, 3)  # plotting FM (Frequency Modulated) signal
plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('FM Signal')

plt.tight_layout()
plt.show()
```
