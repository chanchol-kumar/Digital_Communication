```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
fs = 1000              # Sampling frequency
fc = 50                # Carrier frequency
Am = 1                 # Modulation depth
Ac = 1                 # Carrier amplitude
t = np.arange(0, 1, 1/fs)  # Time vector
message_signal = Am * np.sin(2 * np.pi * 10 * t)  # Information signal

# Phase modulation
kf = 2 * np.pi * 5        # Modulation index (sensitivity)
phase_modulated_signal = Ac * np.cos(2 * np.pi * fc * t + kf * message_signal)

# Plotting
plt.figure(figsize=(8, 12))

plt.subplot(3, 1, 1)
plt.plot(t, message_signal)
plt.title('Message Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.plot(t, Ac * np.cos(2 * np.pi * fc * t))
plt.title('Carrier Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)
plt.plot(t, phase_modulated_signal)
plt.title('Phase Modulated Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.suptitle('Phase Modulation Example')
plt.show()
```
