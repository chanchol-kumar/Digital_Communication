```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = float(input('Enter the amplitude = '))
f = float(input('Enter the frequency  = '))
t = np.arange(0, 2, 0.02)  # for a total of 16 samples

x1 = np.ones(len(t))  # generation of an impulse signal
x2 = a * np.sin(2 * np.pi * f * t)  # generation of sine wave
y = x1 * x2  # modulation step

plt.figure(figsize=(8, 12))

plt.subplot(3, 1, 1)  # for impulse signal plot
plt.stem(t, x1)
plt.title('Impulse Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)  # for sine wave plot
plt.plot(t, x2)
plt.title('Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)  # for PAM wave plot
plt.stem(t, y)
plt.title('PAM Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
