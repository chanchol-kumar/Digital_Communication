```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 20, 0.0001)  # sampling at Nyquist rate
c = int(input('Enter Bit Depth Of PCM Coding: '))

part = np.arange(-1, 1.1, 0.1)  # A quantization partition defines several contiguous ranges
codebook = np.arange(-1, 1.2, 0.1)  # A codebook tells the quantizer which common value to assign to inputs

msg = np.cos(t)
quants = np.digitize(msg, part) - 1
ybin = [format(q, f'0{c}b') for q in quants]

plt.figure(figsize=(8, 12))

plt.subplot(3, 1, 1)
plt.plot(t, msg)
plt.title('Message Signal')

plt.subplot(3, 1, 2)
plt.plot(t, quants)
plt.title('Quantized Signal')

plt.subplot(3, 1, 3)
plt.plot(t, quants)
plt.title('PCM PLOT')

plt.tight_layout()
plt.show()
```
