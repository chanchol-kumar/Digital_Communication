```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
fs = 1000             # Sampling frequency
fc = 50               # Carrier frequency
M = 4                 # Number of symbols
Am = 1                # Modulation depth
t = np.arange(0, 1, 1/fs)  # Time vector
message_symbols = np.random.randint(0, M, size=len(t))  # Random symbol sequence

# Constellation mapping for 4-QAM
constellation = np.array([1 + 1j, -1 + 1j, -1 - 1j, 1 - 1j])

# QAM modulation
qam_modulated_signal = constellation[message_symbols]

plt.figure(figsize=(8, 12))

plt.subplot(3, 1, 1)
plt.stem(t, message_symbols)
plt.title('Message Symbols')
plt.xlabel('Time')
plt.ylabel('Symbol Index')

plt.subplot(3, 1, 2)
plt.plot(np.real(constellation), np.imag(constellation), 'o')
plt.title('4-QAM Constellation')
plt.xlabel('In-phase')
plt.ylabel('Quadrature')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, np.real(qam_modulated_signal), 'b', t, np.imag(qam_modulated_signal), 'r')
plt.title('4-QAM Modulated Signal (In-phase and Quadrature)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend(['In-phase', 'Quadrature'])

plt.tight_layout()
plt.show()
```
